```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('cleaned_articles_mass.csv')
```

```python
df['position_section'] = df['position_section'].replace('G', 'LIVING ARTS')
```

```python
# graphing number of articles by sections
from matplotlib.ticker import FuncFormatter

def format_func(value, tick_number):
    if value >= 1000:
        value /= 1000
        return f"{value}K"
    else:
        return int(value)

# Grouping the DataFrame by 'position_section' and counting the number of articles in each group
articles_by_position = df['position_section'].value_counts()
articles_by_position = articles_by_position.drop(['TAB2', 'unknown', 'TABM','PROJECT','ADVSUP'], errors='ignore')
colors = ['#3fd0d8']

plt.figure(figsize=(10, 6))
ax = articles_by_position.plot(kind='bar')
plt.xlabel('Position Section', fontdict={'fontsize': 14, 'fontweight': 'bold', 'fontfamily': 'Times New Roman'})
plt.ylabel('Number of Articles', fontdict={'fontsize': 14, 'fontweight': 'bold', 'fontfamily': 'Times New Roman'})
plt.title('Number of Articles by Section', fontdict={'fontsize': 16, 'fontweight': 'bold', 'fontfamily': 'Times New Roman'})
plt.xticks(rotation=90, fontfamily='Times New Roman')
plt.yticks(fontfamily='Times New Roman')  # Rotate x-axis labels for better readability and set font family for x-axis labels
plt.tight_layout()

# Customize the y-axis labels
ax.yaxis.set_major_formatter(FuncFormatter(format_func))

# Save the plot as an image
plt.savefig('number_articles_by_section.png', transparent = True)

# Display the plot
plt.show()
```

```python
from matplotlib.ticker import FuncFormatter, MultipleLocator

def format_func(value, tick_number):
    if value >= 1000:
        value /= 1000
        return f"{value}K"
    else:
        return int(value)

# Grouping the DataFrame by 'position_section' and counting the number of articles in each group
articles_by_sub = df['position_subsection'].value_counts()
articles_by_sub = articles_by_sub.drop(['unknown', 'Living Arts', 'G', 'Business','AdvSup', 'RealEstate', 'Diversity','Metro', 'Editorial',' Opinion'], errors='ignore')
colors = ['#3fd0d8']
#'#f6cb02', '#ff0cd1'

plt.figure(figsize=(10, 6))
ax = articles_by_sub.plot(kind='bar')
plt.xlabel('Position Sub Section', fontdict={'fontsize': 14, 'fontweight': 'bold', 'fontfamily': 'Times New Roman'})
plt.ylabel('Number of Articles', fontdict={'fontsize': 14, 'fontweight': 'bold', 'fontfamily': 'Times New Roman'})
plt.title('Number of Articles by Sub Section', fontdict={'fontsize': 16, 'fontweight': 'bold', 'fontfamily': 'Times New Roman'})
plt.xticks(rotation=90, fontfamily='Times New Roman')
plt.yticks(fontfamily='Times New Roman')  # Rotate x-axis labels for better readability and set font family for x-axis labels
plt.tight_layout()

# Customize the y-axis labels
ax.yaxis.set_major_formatter(FuncFormatter(format_func))
ax.yaxis.set_major_locator(MultipleLocator(500))
# Save the plot as an image
plt.savefig('number_articles_by_subsection.png', transparent = True)

# Display the plot
plt.show()
```
