```python
'''
# Function to clean the tract values
def clean_tract(tract):
    return tract.strip("[]' ")

# Applying the function to the 'Tracts' column
bostonGlobe_2010['Tracts'] = bostonGlobe_2010['Tracts'].apply(clean_tract)

'''

# Removing the 2020 census column tract, "Tracts"
bostonGlobe_2010.drop(columns =['Tracts'], inplace = True)


# Printing the DataFrame to verify
#print(bostonGlobe_2010[['Tracts']].head())
```

```python
# Renaming "tract_2010" column 
bostonGlobe_2010.rename(columns = {"tract_2010": "Tract"}, inplace = True )
```

```python
# Loading and examining census 2010 data
census_tract_data = pd.read_csv("census2010.csv")
census_tract_data.head()
```

```python
# Standardizing the tract codes in the census-tract-data to 6 digits
census_tract_data['tract'] = census_tract_data['tract'].astype(str).str.zfill(6)

# Standardizing the tract codes in the bostonGlobe_2010 to 6 digits
bostonGlobe_2010['Tract'] = bostonGlobe_2010['Tract'].astype(str).str.zfill(6)
```

```python
bostonGlobe_2010.head()
```

```python
# Merging the two datasets based on the standardized tract codes
merged_data = pd.merge(bostonGlobe_2010,census_tract_data, left_on='Tract', right_on='tract', how='outer')
```

```python
# Displaying the summary of the merged dataset 
merged_data.info()
```

```python
# Displaying first five rows of dataframe to inspect 
merged_data.head()
```

```python
# List of unique tract codes in the dataset. 
merged_data.tract.unique()
```

```python
# Renaming columns in the dataframe for better readability.
merged_data.rename(columns={"P003001": "T_Total", 
                            "P003002": "T_White alone",
                            "P003003": "T_Black or African American alone", 
                            "P003004": "T_American Indian and Alaska Native alone", 
                            "P003005": "T_Asian alone", 
                            "P003006": "T_Native Hawaiian and Other Pacific Islander alone", 
                            "P003007": "T_Some Other Race alone", 
                            "P003008": "T_Population of two or more races", 
                            "P004001": "T_Total Hispanic or Latino origin",
                            "P004002": "T_Not Hispanic or Latino",
                            "P004003": "T_Hispanic or Latino"}, inplace=True)
```

```python
# new dataframe with article title, tract number and demogrpahic
df = merged_data[['year','hl1_x','tract', 'T_White alone',
                  'T_Black or African American alone',
                  'T_Hispanic or Latino', 
                  'T_Asian alone', 
                  'T_American Indian and Alaska Native alone', 
                  'T_Native Hawaiian and Other Pacific Islander alone', 
                  'T_Some Other Race alone', 
                  'T_Population of two or more races']].copy()
```

```python
# Displaying first five rows of dataframe to inspect 
df.head()
```

```python
# Counting occurrences of each tract and maping these counts to a new 'Article_Count' column.
tract_counts = df['tract'].value_counts()
df['Article_Count'] = df['tract'].map(tract_counts)
```

```python
# Dataframe with tract code and article count
count_df = df[['year','tract', 'Article_Count']].copy()
```

```python
# Droping duplicates 
count_df = count_df.drop_duplicates(subset=['tract'])
```

```python
# Droping missing values 
count_df = count_df.dropna(subset=['tract'])
```

```python
count_df.head()
```

```python
# Converting dataframe to csv file for article count and tract 
# ---> above steps were conducted for all years and merged together and used to visualize article density for each tract over the years.

count_df.to_csv('count_2010.csv', index=False)
```

```python
# Dataframe with tract number and demographic stats for each tract 
tract_race_df = df[['tract', 'T_White alone',
                  'T_Black or African American alone',
                  'T_Hispanic or Latino', 
                  'T_Asian alone', 
                  'T_American Indian and Alaska Native alone', 
                  'T_Native Hawaiian and Other Pacific Islander alone', 
                  'T_Some Other Race alone', 
                  'T_Population of two or more races']].copy()
```

```python
# removing duplicates 
tract_race_df.drop_duplicates(inplace=True)
tract_race_df.head()
```

```python
tract_race_df.dropna(how='all', inplace=True)
```

```python
max_race_df = pd.DataFrame(tract_race_df)

# Converting all columns to numeric
max_race_df.set_index("tract", inplace=True)
new_df_numeric = max_race_df.apply(pd.to_numeric, errors='coerce')

# Identifying the highest race for each tract and adding a new column
max_race_df["T_majority_race"] = new_df_numeric.idxmax(axis=1)

# Displaying the updated dataframe
max_race_df.reset_index(inplace=True)
max_race_df.head()
```

```python
# Dataframe with number of articles and majority race for each tract 
df = pd.merge(count_df, max_race_df[['tract', 'T_majority_race']], on='tract')

# Displaying the updated DataFrame
df.head()
```

```python
# Grouping the data by 'majority_race' and summing up the 'Article_Count'
article_counts_by_race = df.groupby('T_majority_race')['Article_Count'].sum()

# Converting the result to a dataframe 
article_counts_df = article_counts_by_race.reset_index()

# Renaming columns for better readability
article_counts_df.columns = ['T_Majority Race', 'Article Count']

# Printing the result
print("Number of articles per race:")
print(article_counts_df.to_string(index=False))
```

```python
# droping columns not needed 
merged_data.drop(columns=['NAME', 'state', 'county'], inplace=True)
```

```python
# Select specific columns from df2 to merge into df1
columns_to_merge = ['tract', 'Article_Count', 'T_majority_race']
final_df = merged_data.merge(df[columns_to_merge], on='tract', how='left')
```

```python
# Saving cleaned dataframe to a new CSV file
final_df.to_csv('race_2010.csv', index=False)
```
